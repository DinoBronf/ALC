### Pregunta 1 - ¿Por que la matriz A definida en (4) es equivalente a $p \ \mathbf{W} \ \mathbf{D} + e \ z^t$ ? Justificar.

# Enunciado 1

**Para demostrar que la matriz $A$ es equivalente a $p W D + e z^t$,
primero vamos a recordar qué es cada cosa.**

Los componentes de la matriz $A$ por definición son:
$$a_{i,j} = \begin{cases}
    \begin{aligned}
    \dfrac{1-p}{n} + \dfrac{p \cdot w_{i,j}}{c_j} & \text{si } c_j \neq 0 \\
    \dfrac{1}{n} & \text{si } c_j = 0
    \end{aligned}
\end{cases}$$

donde:

-   La matriz $W$ es una matriz cuyos componentes son $0$ o $1$,
    dependiendo de si la página $j$ tiene un enlace a la página $i$.
    Entonces,
    $w_{ij} = \begin{cases} 0 & \text{si no hay enlace} \\ 1 & \text{si hay enlace} \end{cases}$.
    La diagonal de esta matriz es de $0$s, ya que no cuenta los
    autoenlaces.

-   Con la matriz $W$ se construye $c_j$, siendo
    $c_j = \sum_{i=1}^{n} w_{ij}$.

-   La matriz $D$ es una matriz diagonal, donde
    $D_{jj} = \begin{cases} \dfrac{1}{c_j} & \text{si } c_j \neq 0 \\ 0 & \text{si } c_j = 0 \end{cases}$.

-   $p$ es un número entre $(0,1)$.

-   El vector $e$ es un vector columna de unos.

-   El vector $z^t$ es un vector donde
    $z_j = \begin{cases} \dfrac{1-p}{n} & \text{si } c_j \neq 0 \\ \dfrac{1}{n} & \text{si } c_j = 0 \end{cases}$.

**Para probar la igualdad vamos a resolver paso a paso $p W D + e z^t$,
comenzando por $p W D$:**  
Al hacer $p W$ siendo $p$ una variable, nos queda que la matriz $p W$
es: $p \cdot w_{i,j}$.  
Luego, al hacer el producto de matrices $p W D$ nos va a quedar una
nueva matriz con coordenadas $\dfrac{p \cdot w_{i,j}}{c_j}$ y una
diagonal de 0s, ya que la matriz $p W$ posee una diagonal de 0s. 
Entonces en este punto tenemos una matriz que luce así:

$$
{
\begin{bmatrix}
    0 & \frac{p \cdot w_{1,2}}{c_2} & \frac{p \cdot w_{1,3}}{c_3} & \cdots & \frac{p \cdot w_{1,n}}{c_n} \\
    \frac{p \cdot w_{2,1}}{c_1} & 0 & \frac{p \cdot w_{2,3}}{c_3} & \cdots & \frac{p \cdot w_{2,n}}{c_n} \\
    \frac{p \cdot w_{3,1}}{c_1} & \frac{p \cdot w_{3,2}}{c_2} & 0 & \cdots & \frac{p \cdot w_{3,n}}{c_n} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    \frac{p \cdot w_{n,1}}{c_1} & \frac{p \cdot w_{n,2}}{c_2} & \frac{p \cdot w_{n,3}}{c_3} & \cdots & 0
\end{bmatrix}
}
$$ 
Aclaración: En este caso estamos asumiendo que $c_{j}$ es distinto de 0. En el caso en el que sea 0, la coordenada sería igual a 0

**Para continuar, vamos a hacer el producto de $e z^t$:**  
Al estar multiplicando un vector columna de unos por un vector fila nos va a quedar una matriz de la pinta:
$$\begin{bmatrix}
    z_{11} & z_{12} & \cdots & z_{1n} \\
    z_{11} & z_{12} & \cdots & z_{1n} \\
    \vdots & \vdots & \ddots & \vdots \\
    z_{11} & z_{12} & \cdots & z_{1n}
\end{bmatrix}
$$  
Que tiene coordenadas de la forma: $ z_1,j = \begin{cases} \frac{1-p}{n} & \text{si } c_j \neq 0 \\ \frac{1}{n} & \text{si } c_j = 0 \end{cases} $

**Por último, sumamos las dos matrices calculadas, y nos queda algo de la forma:** 
$${
\begin{bmatrix}
    z_{11} & \frac{p \cdot w_{1,2}}{c_2}+z_{12} & \frac{p \cdot w_{1,3}}{c_3}+z_{13} & \cdots & \frac{p \cdot w_{1,n}}{c_n}+z_{1n} \\
    \frac{p \cdot w_{2,1}}{c_1}+z_{21} & z_{22} & \frac{p \cdot w_{2,3}}{c_3}+z_{23} & \cdots & \frac{p \cdot w_{2,n}}{c_n}+z_{2n} \\
    \frac{p \cdot w_{3,1}}{c_1}+z_{31} & \frac{p \cdot w_{3,2}}{c_2}+z_{32} & z_{33} & \cdots & \frac{p \cdot w_{3,n}}{c_n}+z_{3n} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    \frac{p \cdot w_{n,1}}{c_1}+z_{n1} & \frac{p \cdot w_{n,2}}{c_2}+z_{n2} & \frac{p \cdot w_{n,3}}{c_3}+z_{n3} & \cdots & z_{nn}
\end{bmatrix}
}$$

Llamemos a esta matriz "B". Vemos que las coordenadas de B dependen también de cj, y nos queda:

$$ 
b_{i,j} = \begin{cases}
    \begin{aligned}
    \dfrac{1-p}{n} + \dfrac{p \cdot w_{i,j}}{c_j} & \text{si } c_j \neq 0 \\
    \dfrac{1}{n} & \text{si } c_j = 0
    \end{aligned}
\end{cases}
$$ 
**Con esto finalmente llegamos a que la nueva matriz B,
proveniente de $p W D + e z^t$ es igual a la matriz A, y queda probado
que A = $p W D + e z^t$**

# Enunciado 2
**Para demostrar que la matriz ($I - p W D$) tiene factorizacion LU vamos a probar primero que la traspuesta de esta admite LU, y que consecuentemente esta también. Para el primer paso vamos a utilizar dos propiedades sobre las matrices estrictamente diagonal dominantes:**
Una matriz es estrictamente diagonal dominante cuando cumple que;
$$
\sum_{{\substack{j=1 \\ j \neq i}}}^{n} |a_{i,j}| < |a_{i,i}|
$$

- Si A es EDD entonces, trabajando con aritmética exacta, el proceso
de eliminación de Gauss no produce pivots nulos.
- Sea A ∈ $R^(nxn)$ EDD entonces A es invertible.

Con estas propiedades queda claro que si podemos probar que la matriz ${(I-p W D)}^{t}$ es EDD, entonces se puede garantizar la existencia de su factorización LU.

**Primero que nada vamos a analizar como queda la matriz ($I- p W D$) una vez hecha la resta.:**

Utilizamos que ya calculamos previamente $p W D$ y entonces nos queda:
$$
\mathbf{A} = 
\begin{bmatrix}
1 & \frac{-p \cdot w_{12}}{c_2} & \frac{-p \cdot w_{13}}{c_3} & \cdots & \frac{-p \cdot w_{1n}}{c_n} \\
\frac{-p \cdot w_{21}}{c_1} & 1 & \frac{-p \cdot w_{23}}{c_3} & \cdots & \frac{-p \cdot w_{2n}}{c_n} \\
\frac{-p \cdot w_{31}}{c_1} & \frac{-p \cdot w_{32}}{c_2} & 1 & \cdots & \frac{-p \cdot w_{3n}}{c_n} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
\frac{-p \cdot w_{n1}}{c_1} & \frac{-p \cdot w_{n2}}{c_2} & \frac{-p \cdot w_{n3}}{c_3} & \cdots & 1
\end{bmatrix}
$$
**Aclaración: en este caso, al igual que en el enunciado 1, estamos suponiendo que $c_{j}$ es distinto de 0.**

De esta manera, la traspuesta de esta luce así: 
$$
\mathbf{A}^T = 
\begin{bmatrix}
1 & \frac{-p \cdot w_{21}}{c_1} & \frac{-p \cdot w_{31}}{c_1} & \cdots & \frac{-p \cdot w_{n1}}{c_1} \\
\frac{-p \cdot w_{12}}{c_2} & 1 & \frac{-p \cdot w_{32}}{c_2} & \cdots & \frac{-p \cdot w_{n2}}{c_2} \\
\frac{-p \cdot w_{13}}{c_3} & \frac{-p \cdot w_{23}}{c_3} & 1 & \cdots & \frac{-p \cdot w_{n3}}{c_3} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
\frac{-p \cdot w_{1n}}{c_n} & \frac{-p \cdot w_{2n}}{c_n} & \frac{-p \cdot w_{3n}}{c_n} & \cdots & 1
\end{bmatrix}
$$
**Para que la matriz  ${(I-p W D)}^{t}$ sea EDD, se tiene que cumplir que la suma de las coordenadas de cada columna (sin incluir la diagonal) sea menor a 1.**
Sabemos que la sumatoria de las filas de la matriz ($W D$) es igual a 1, ya que estamos sumando todos los links que llevan a la página 1 y dividiendo estos por la cantidad de links que hay. De esta manera, al analizar las columnas de ${(I-p W D)}^{t}$ vemos que resulta EDD, ya que p varía entre (0,1) pero nunca toma estos valores, por lo que nunca la sumatoria de las coordenadas de las columnas (excluyendo a la coordenadad diagonal) va a ser mayor o igual a 1, ya que 
$$
\sum_{{\substack{j=1 \\ j \neq i}}}^{n} |p * a_{i,j}| = p * 1 
$$
con p < 1 y la coordenada $a_{i,j} = \frac{w_{i,j}}{c_j}$.
**Así demostramos que la traspuesta de la matriz ($I - p W D$) es efectivamente EDD**, entonces (por Proposición 4.3.del apunte), el proceso
de eliminación de Gauss no produce pivots nulos, por lo que **admite factorización LU.** Y a su vez (por Proposición 4.2. del apunte) como es EDD entonces **es invertible**. 
**Ahora bien, sabemos que si esta matriz traspuesta tiene LU y es inversible, todas sus submatrices principales tienen un determinante distinto de 0. Probar que la matriz deseada admite una factorización LU entonces resulta fácil, ya que, siguiendo la propiedad**:

 $ \large det(A) = det({A}^{t}) $
Se puede entender entonces que los determinantes de las submatrices de la matriz ($I - p W D$) van a ser distintos de cero, probando así, **que la matriz ($I - p W D$) admite una factorización LU**.


**Ahora se busca analizar si la matriz A = ($I - p W D$) está bien condicionada y como afecta el parametro p a esto. Para eso vamos a hacer un analisis sobre la condicion de esta matriz:**

Sabemos que Cond($I - p W D$) = ||($I - p W D$)|| * ||$(I - pWD)^{-1}$||.
- Vemos que si p tiende a 0 ocurre que
$$ \lim_{{p \to 0}} \left\| (I - pWD) \right\| \approx 1 $$
$$ \lim_{{p \to 0}} \left\| (I - pWD)^{-1} \right\| \approx 1 $$
y se puede ver claramente que entonces la condicion de esta matriz va a tender a 1:
$$ \text{Cond}(I - pWD) \approx 1 $$
